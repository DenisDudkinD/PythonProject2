```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from src.domain.book import Book

df = pd.read_json('books.json')
min_thresh = 100
median = df.groupby('genre')['ratings_count'].median()
mean = df.groupby('genre')['average_rating'].mean()
mean_all_books = df['average_rating'].mean()
weighted_rating = (median/(median+min_thresh))*mean+(min_thresh/(median+min_thresh))*3.255

plt.bar(df['genre'].unique(),weighted_rating)
plt.xlabel('Genres')
plt.ylabel('Bayesian Rating')
plt.title('Bayesian Rating of Books by Genre')
```
